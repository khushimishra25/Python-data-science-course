```python
# to change column name
df.rename(mapper={"make":"brand"},axis=1,inplace=True)      # by default searching in index row wise, axis=1 makes it column, inplace true to make the actual change in data#
# no change in the actual file
```

```python
df.head()      
```

```python
# dropping the column
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.head()
```

```python
df.replace('alfa-romero','alfa-romeo',inplace=True)
```

```python
df
```

```python
df.replace('?',np.nan, inplace=True)        #removes all ? by np.nan 
```

```python
df
```

```python
#df.dropna() removes all entry with NaN
df.dropna(subset=['price'],inplace=True) #removes all missing price values
```

```python
df.info()
```

```python
df.isnull().sum()       # count the number of missing data in each column
```

```python
df['normalized-losses'].astype('float').sum()       #changes datatype to float and sums it
```

```python
mnl = df['normalized-losses'].astype('float').mean()        
```

```python
df['normalized-losses'].replace(np.nan,mnl,inplace=True)        # replaces nan values with mean values
```

```python
df.head()
```

```python
mbore = df['bore'].astype('float').mean()  
df['bore'].replace(np.nan,mbore,inplace=True)
df
```

```python
ms = df['stroke'].astype('float').mean()  
df['stroke'].replace(np.nan,ms,inplace=True)
df
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
```

```python
df.isnull().sum()
```

```python
# to save data
df.to_csv("automobile_cleaned.csv",index=False)
```
